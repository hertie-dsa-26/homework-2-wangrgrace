## Question 2: Pointers and References (1 point)

Examine the following Python code carefully:

```python
def check_array(input):
    for idx in range(len(input)):
        if input[idx][0] == "1":
            input[idx] = None
    return input

original_array = ["1_3", "5_2"]
new_array = check_array(original_array)

print(original_array)
print(new_array)
```

What does this code output? Explain **why** it produces that output, using the concepts of pointers and references in Python.

````python
def check_array(input):
    for idx in range(len(input)):
        if input[idx][0] == "1":
            input[idx] = None
    return input

original_array = ["1_3", "5_2"]
print(original_array)

new_array = check_array(original_array)
print(new_array)
````

Literally, the function *check_array()* loops through each of the elements in a list. It outputs *"None"* if the first character of the element is "1", and the original element otherwise.

After running this code as is, it yields *[None, '5_2']* for both *print(original_array)* and *print(new_array)*. This seems strange at first because while the result for print(new_array) is as expected, one would expect print(original_array) to yield *['1_3', '5_2']*. *original_array* is a pointer to a list object stored in the computer's memory, and its pointer value (that which it references) changes after this line of code:

*new_array = check_array(original_array)*

Because this line occurs before *print(original_array)* in the original code, the pointer value of original_array has already passed through the function, hence yielding *[None, '5_2']*. I've changed the order of the code above to avoid this issue. This allows us to display *original_array* as is before applying the function.

Alternatively, you could add the line *input = input.copy()* within the function to preserve the original pointer.

## Question 3: Algorithm Analysis (2 points)

Consider the following algorithm, which finds the contiguous subarray with the largest sum:

```python
max_sum = None
for i in range(len(array)):
    sum_subarray = 0
    for j in range(i, len(array)):
        sum_subarray += array[j]
        if max_sum is None or max_sum < sum_subarray:
            max_sum = sum_subarray
print(max_sum)
```

Analyze the efficiency of this algorithm:

1. Is this algorithm polynomial? Justify your answer.
2. What is the runtime of this algorithm in **Big-O** notation? Show your reasoning.

````python
array = [1, -3, 5, 7, -9]

max_sum = None
for i in range(len(array)):
    sum_subarray = 0
    for j in range(i, len(array)):
        sum_subarray += array[j]
        if max_sum is None or max_sum < sum_subarray:
            max_sum = sum_subarray
print(max_sum)
````

1) Yes. In general, when one loop is nested inside another and each iterates An + B times (where A and B are numerical constants that don't contain n), the number of iterations simplifies to n². The highest order term in this expression is thus *n²*, a polynomial expression.

2) The loop *for i in range(len(array))* runs n times, while the loop *for j in range(i, len(array))* runs (n-i) times, where i is any value from 0 to (n-1). As follows, the code runs *1 + 2 +...+ (n-1) + n* times, which is equal to *n(n+1)/2*. The complete expression is *n²/2 + n/2*, and after simplification, the runtime of the algorithm is *O(n²)* in Big-O notation.
